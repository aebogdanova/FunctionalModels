На первый взгляд кажется, что выделение трех типов отношений между парами слов, фраз или предложений лучше всего описывает все возможные логические связи: одно выражение может либо следовать из другого, либо противоречить ему, либо никак к нему не относиться. Такая модель обычно обладает явным преимуществом перед моделью, которая описывается двумя типами отношений (одно выражение либо следует из другого, либо не следует). Даже интуитивно понятно, что пространство признаков, по которым человек строит выводы о связи двух слов или предложений, на самом деле шире. Существует четкое разграничение между противоречием и отсутствием следствия: если первое определяет априори ложное высказывание по отношению к истинному, то второе в принципе не может дать определения в категориях истинности/ложности. Например, для предпосылки (1) гипотезы (2a) и (2b) описывают разные типы логической связи: если село отрицает город (противопоставление по количеству жителей и статусу населенного пункта), то Москва - нет (но и не следует из него: Москва - город, но город - не обязательно Москва). 

(1) Я приехал в родной город.<br>
(2a) Я приехал в родное село.<br>
(2b) Я приехал в Москву.


Здесь следует отдельно отметить, что меньшее количество лейблов не облегчает задачу определения типа отношений. Какое, например, отношение можно установить между словами "человек" и "животное"? С одной стороны, биологически человек относится к царству животных, тогда из "человека" следует "животное". Однако все же чаще мы эти слова противопоставляем друг другу (ср., например, "Он ведет себя как животное."), когда хотим подчеркнуть принципиальное отличие человека от других представителей животного мира. Такие же случаи неоднозначного определения типа отношений можно наблюдать у синонимов, особенно если они описывают частичное совпадение дифференциальных признаков сигнификата, а не полное. Например, "плакать" следует из "рыдать" по определению, т.к. и "рыдать", и "плакать" эквивалентны "проливать слезы", однако если в выражении ставится акцент на интенсивности действия (признак, по которому синонимы различаются), то следствие пропадает: "Она не просто плакала, она рыдала". Еще один пример -- это стилистические синонимы, для которых определение типа отношений тоже может быть затруднено: денотат для слов "глаза" и "очи" совпадает (то есть формально они эквивалентны и следуют друг из друга), однако употребление их в речи может демонстрировать противоречие: "Глаза у нее были большие, темно-карие -- не глаза, а очи".

Разбиение на четыре типа отношений (тождественность, включение, обратное включение, не-следствие) по большей степени всего лишь расширяет понятие следствия: два выражения могут быть эквивалентны (то есть каждое из выражений может быть следствием из другого, такие выражения еще называют парафразами), либо из двух выражений только одно следует из другого (тогда если из предпосылки следует гипотеза, то говорят о включении, а если из гипотезы следует предпосылка, то об обратном включении). Такое расширение пространства отношений, на самом деле, добавляет небольшую, но все же существенную деталь в определение логических связей между фразами и тем самым может несколько облегчить выбор лейбла. Например, если бы мы определяли отношения между предпосылкой (3) и гипотезой (4) и руководствовались только тремя типами отношений, то, скорее всего, затруднились бы дать однозначный ответ: с одной стороны, подходит "contradiction", так как Палата представителей - это одна из двух палат Конгресса, и если мы опираемся на максиму количества (по Грайсу), то Палата представителей в данном контексте противопоставлена Конгрессу; с другой стороны, если у нас нет уверенности в кооперативности участников коммуникации, то мы поставили бы лейбл "neutral", так как если Палата представителей (нижняя палата Конгресса) утвердила план, то это еще не означает, что Сенат (верхняя палата Конгресса) его не утвердил. Если же пространство отношений расширить, то неоднозначность легко снимается: очевидно, что (4) обратно включает (3): если Конгресс утвердил план, значит и Палата представителей его утвердила.

(3) Палата представителей утвердила план предотвращения "фискального обрыва".<br>
(4) Конгресс США утвердил план по предотвращению "фискального обрыва".

Расширение наименований отношений за счет включения конверсациональных импликатур кажется полезным, так как этот компонент высказывания часто играет большую роль в коммуникативном акте. Однако есть опасения, что, во-первых, создать такой корпус не удастся в силу того, что конверсациональные импликатуры сильно завязаны на контекст, знания о внешнем мире, цели участников коммуникативного акта и степень их кооперации (а потому лейблы разных аннотаторов могут быть совершенно разными, но при этом одинаково верными для каждого, например, для предложения (5) импликатурой будет и (6a), и (6b), а нам бы хотелось избегать неоднозначности). Неопределенность и отменяемость конверсациональных импликатур делают практически невозможным создание такого корпуса.

(5) Вы настоящие друзья. <br>
(6а) Спасибо, что помогли мне. <br>
(6b) Вы оставили меня в беде. (Принцип иронии)

Более реальным кажется создание корпуса пресуппозициональных отношений. Во-первых, для пресуппозиции есть измеримое свойство: она сохраняется под операторами, отменяющими следствие. Во-вторых, семантические пресуппозиции часто связаны с конкретными языковыми выражениями и конструкциями (триггерами), а значит, могут быть выведены из высказывания лингвистически. ### Задание 4

Для оценки комитмента говорящего авторы CommitmentBank используют 7-интервальную шкалу Ликерта: аннотаторы оценивают степень комитмента говорящего к содержанию комплемента по шкале от -3 до +3. Такая оценка кажется вполне оптимальной для поставленной задачи по нескольким причинам. 

Во-первых, с ее помощью можно делать выводы о фактивности глаголов и ее корреляции с оценками комитмента. Например, интуитивно глагол "подозревать" кажется нефактивным: "Я подозреваю [= "полагаю", "считаю" - нефактивные глаголы мнения], что он прав". Изучая поведение этого глагола под действием операторов, отменяющих следствие, скорее всего, мы увидим, что глагол иногда демонстрирует высокую степень фактивности, особенно под действием отрицания: "Я и не подозревал [= "не знал" - фактивный глагол знания], что он прав", то есть оценка комитмента в каких-то случаях будет высокой. Анализ корреляции фактивности глагола и оценок комитмента (в статье - с использованием статистической модели со смешанными эффектами) будет эффективнее, если будут использоваться оценки широкого диапазона (разброс оценок от -1 до 1 слишком мал для корректного статистического анализа). 

Во-вторых, при использовании шкалы Ликерта может возникнуть проблема, связанная с поведением аннотаторов: они могут быть склонны избегать крайних оценок. Тогда 7-интервальная шкала выглядит даже оптимальней, чем, например, 5-интервальная, потому что, если крайние оценки (-3 и +3) будут игнорироваться аннотаторами, останется еще пять значений, которых должно быть достаточно для исследования.

На мой взгляд, оценивать комитмент субъекта пропозиционального глагола в исследовании пропозиции и фактивности глаголов не имеет смысла, прежде всего потому, что комитмент субъекта к содержанию пропозиции будет определяться самим значением глагола: например, в предложениях (1) и (2) субъект пропозициального предиката в первом случае твердо уверен, что пропозиция "отец прав" истинна, во втором случае его уверенность выражена меньше, и различие в оценках прямо зависит от выбора соответствующего глагола (выражающего разную степень уверенности в своем мнении).

(1) "Вася верит, что отец прав."<br>
(2) "Вася считает, что отец прав."

Тогда если поставить глаголы под операторы, отменяющие следствие, значение глаголов будет меняться, соответственно, будет меняться комитмент субъекта: в предложении (3) и (4) пропозиция "отец прав" субъектом отрицается.

(3) Вася не верит, что отец прав.<br>
(4) Вася не считает, что отец прав.

Такая оценка комитмента ничего не будет говорить о содержании самой пропозиции, и сделать выводы о ее проективности мы не сможем. Изучая прагматическую сторону высказывания, компоненты его смысла (пресуппозицию, импликатуры), мы должны опираться прежде всего на говорящего как на субъекта этого высказывания.


### Задание 5 (вариант 1)

Для автоматического извлечения стимулов и их аннотации потребуется:
- морфологический анализатор: для извлечения пропозициционального глагола (его времени, леммы) и субъекта действия (лица, числа, леммы),
- синтаксический парсер: для извлечения комплемента и операторов, отменяющих следствие, в сферу действия которых попадает пропозициональный предикат.

В данном варианте решения задачи используется библиотека для русского языка natasha.

```python
from natasha import Doc, Segmenter, MorphVocab, NewsEmbedding, NewsMorphTagger, NewsSyntaxParser
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
```

```python
# function for extracting annotation

def extract_annotations(text, genre='unknown'):
    '''takes two arguments: corpus of interest (should be plain text in string format) and its genre (default is 'unknown')'''
    
    # morphology tagging and syntax parsing of text given
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.parse_syntax(syntax_parser)
    
    data = {}       # dictionary of annotations: 
                    # context: preceding context of the target sentence (2 sentences)
                    # target: sentence of interest which contains a clause-embedding predicate under an entailment canceling operator
                    # verb: clause-embedding predicate
                    # embedding: type of entailment canceling operator
                    # mat_tense: tense of the matrix verb
                    # mat_subj_lemma: lemma of the matrix verb subject
                    # mat_subj_per: person of the matrix verb subject
                    # mat_subj_num: number of the matrix verb subject
                    # genre: genre of corpus given as an argument of the function
    id_number = 0   # id of annotation (key of dictionary of annotations)

    MODALS = ['хотеть', 'желать', 'мочь', 'надо', 'нужный', 'нужно', 'нельзя',\
              'можно', 'должен', 'должный', 'необходимо', 'необходимый']
    
    for sent in doc.sents:
        for token in sent.tokens:
            
            # search clausal complement
            if token.rel == 'ccomp':
                ccomp = token
                
                # search predicate (indicative verb)
                try:
                    for token in sent.tokens:
                        if token.id == ccomp.head_id and token.pos == 'VERB' and token.feats['Mood'] == 'Ind':
                            predicate = token
                            predicate.lemmatize(morph_vocab)
                        
                            # for russian some modals are parsed as predicate (indicative verb),
                            # so we should define them correctly and add to annotations with embedding type 'modal'
                            if predicate.lemma in MODALS:
                                for token in sent.tokens:
                                    if token.head_id == predicate.id and token.pos == 'VERB':
                                        verb = token
                                        verb.lemmatize(morph_vocab)
                                        # search subject
                                        for token in sent.tokens:
                                            if token.head_id == predicate.id and token.rel == 'nsubj':
                                                mat_subject = token
                                                mat_subject.lemmatize(morph_vocab)
                                                # save to dictionary with appropriate annotation
                                                try:
                                                    data[id_number] = {
                                                        'context': str(
                                                            doc.sents[doc.sents.index(sent)-2].text\
                                                            +' '+\
                                                            doc.sents[doc.sents.index(sent)-1].text
                                                            ),
                                                        'target': sent.text,
                                                        'verb': verb.lemma,
                                                        'embedding': 'modal',
                                                        'mat_tense': predicate.feats['Tense'],
                                                        'mat_subj_lemma': mat_subject.lemma,
                                                        'mat_subj_per': mat_subject.feats['Person'],
                                                        'mat_subj_num': mat_subject.feats['Number'],
                                                        'genre': genre
                                                        }
                                                    id_number += 1
                                                except KeyError:
                                                    pass               
                            else:
                                # search negation 
                                for token in sent.tokens:
                                    if token.head_id == predicate.id and token.text == 'не':
                                        # search subject
                                        for token in sent.tokens:
                                            if token.head_id == predicate.id and token.rel == 'nsubj':
                                                mat_subject = token
                                                mat_subject.lemmatize(morph_vocab)
                                                # save to dictionary with appropriate annotation
                                                try:                              
                                                    data[id_number] = {
                                                        'context': str(
                                                            doc.sents[doc.sents.index(sent)-2].text\
                                                            +' '+\
                                                            doc.sents[doc.sents.index(sent)-1].text
                                                            ),
                                                        'target': sent.text,
                                                        'verb': predicate.lemma,
                                                        'embedding': 'negation',
                                                        'mat_tense': predicate.feats['Tense'],
                                                        'mat_subj_lemma': mat_subject.lemma,
                                                        'mat_subj_per': mat_subject.feats['Person'],
                                                        'mat_subj_num': mat_subject.feats['Number'],
                                                        'genre': genre
                                                        }
                                                    id_number += 1
                                                except KeyError:
                                                    pass
                                        
                                # search question 
                                for token in sent.tokens:
                                    if token.text == '?':
                                        # search subject
                                        for token in sent.tokens:
                                            if token.head_id == predicate.id and token.rel == 'nsubj':
                                                mat_subject = token
                                                mat_subject.lemmatize(morph_vocab)
                                                # save to dictionary with appropriate annotation
                                                try:                                        
                                                    data[id_number] = {
                                                        'context': str(
                                                            doc.sents[doc.sents.index(sent)-2].text\
                                                            +' '+\
                                                            doc.sents[doc.sents.index(sent)-1].text
                                                            ),
                                                        'target': sent.text,
                                                        'verb': predicate.lemma,
                                                        'embedding': 'question',
                                                        'mat_tense': predicate.feats['Tense'],
                                                        'mat_subj_lemma': mat_subject.lemma,
                                                        'mat_subj_per': mat_subject.feats['Person'],
                                                        'mat_subj_num': mat_subject.feats['Number'],
                                                        'genre': genre
                                                        }
                                                    id_number += 1
                                                except KeyError:
                                                    pass
                                                
                except KeyError:
                    pass
                
                # search conditional
                for token in sent.tokens:
                    # search predicate under conditional
                    if token.id == ccomp.head_id and token.pos == 'VERB' and token.rel == 'advcl':
                        predicate = token
                        predicate.lemmatize(morph_vocab)
                        # search subject
                        for token in sent.tokens:
                            if token.head_id == predicate.id and token.rel == 'nsubj':
                                mat_subject = token
                                mat_subject.lemmatize(morph_vocab)
                                # save to dictionary with appropriate annotation
                                try:
                                    data[id_number] = {
                                        'context': str(
                                        doc.sents[doc.sents.index(sent)-2].text\
                                        +' '+\
                                        doc.sents[doc.sents.index(sent)-1].text),
                                        'target': sent.text,
                                        'verb': predicate.lemma,
                                        'embedding': 'conditional',
                                        'mat_tense': predicate.feats['Tense'],
                                        'mat_subj_lemma': mat_subject.lemma,
                                        'mat_subj_per': mat_subject.feats['Person'],
                                        'mat_subj_num': mat_subject.feats['Number'],
                                        'genre': genre
                                        }
                                    id_number +=1
                                except KeyError:
                                    pass
                
                
                # search modal
                for token in sent.tokens:
                    # search predicate
                    if token.id == ccomp.head_id and token.pos == 'VERB':
                        predicate = token
                        predicate.lemmatize(morph_vocab)
                        # search modal
                        for token in sent.tokens:
                            token.lemmatize(morph_vocab)
                            if token.id == predicate.head_id and token.lemma in MODALS:
                                modal = token
                                # search subject
                                for token in sent.tokens:
                                    if token.head_id == modal.id and token.rel == 'nsubj' or token.rel == 'iobj':
                                        mat_subject = token
                                        mat_subject.lemmatize(morph_vocab)
                                        # save to dictionary with appropriate annotation
                                        try:
                                            data[id_number] = {
                                                'context': str(
                                                doc.sents[doc.sents.index(sent)-2].text\
                                                +' '+\
                                                doc.sents[doc.sents.index(sent)-1].text),
                                                'target': sent.text,
                                                'verb': predicate.lemma,
                                                'embedding': 'modal',
                                                'mat_tense': modal.feats['Tense'],
                                                'mat_subj_lemma': mat_subject.lemma,
                                                'mat_subj_per': mat_subject.feats['Person'],
                                                'mat_subj_num': mat_subject.feats['Number'],
                                                'genre': genre
                                                }
                                            id_number +=1
                                        except KeyError:
                                            pass
                                      
    return data
```
